```python
# In this code, I want to create a pivot table in which all state's mean bed capacity per 1000
# is sorted for all hospital types (ICU, ACUTE, PSYCHIATRIC, OTHER)

import pandas as pd
import numpy as np
df = pd.read_csv('hospital_beds_USA_v1.csv')
pt1 = df.pivot_table(values='beds', index='state', columns='type', aggfunc=[np.mean])

pt1.head()
```
